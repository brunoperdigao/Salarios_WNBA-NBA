# WBNA Salaries

```python
df = pd.read_html("https://www.spotrac.com/wnba/rankings/") # The year 2022 were select in the web page
df = df[0]
df = df.iloc[:, 1:]
df
```

```python
df['Player'] = df['Player'].astype('string')
df['Team'] = df['Player'].str.split(" ")
df['Team'] = [text[-1] for text in df['Team']]
df['Team'] = df['Team'].astype('string')
df['Player'] = [player.replace('  ' + team, '') for player, team in zip(df['Player'], df['Team'])]
df
```

```python
df['cap hit'] = df['cap hit'].astype('string')
df['cap hit'] = df['cap hit'].str.replace('$', '')
df['cap hit'] = df['cap hit'].str.replace(',', '')
df['cap hit'] = df['cap hit'].astype('int64')
df.info()
df
```
